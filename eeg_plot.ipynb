```python
import pandas as pd

dataset=pd.read_csv("user_a_0.csv")

frame = pd.DataFrame(dataset)
#frame
```

```python
col=['AF3 alpha std','AF3 beta std','AF3 delta std','AF3 theta std']
df=pd.DataFrame(frame, columns=col)
```

```python
import matplotlib.pyplot as plt
fig,a =  plt.subplots(4, figsize=(10,6))
fig.tight_layout()

a[0].plot(df['AF3 alpha std'])
a[0].set_title('AF3-ALPHA')
a[1].plot(df['AF3 beta std'])
a[1].set_title('AF3-BETA')
a[2].plot(df['AF3 delta std'])
a[2].set_title('AF3-DELTA')
a[3].plot(df['AF3 theta std'])
a[3].set_title('AF3-THETA')
plt.show()
```
